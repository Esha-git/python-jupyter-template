```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
import matplotlib.pyplot as plt
import numpy as np

# Create x values from 0 to 10
x = np.arange(0, 11)

# Compute y as the square of x
y = x ** 2

# Create the plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, marker='o', linestyle='-', color='b')

# Add title and labels
plt.title('Line Plot of x vs x²')
plt.xlabel('x')
plt.ylabel('x squared')

# Add grid
plt.grid(True)

# Show the plot
plt.show()
```

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'orange']  # Different colors for each bar

# Create bar plot
plt.bar(categories, values, color=colors)

# Add title
plt.title('Category Values Bar Plot')

# Add labels for clarity
plt.xlabel('Categories')
plt.ylabel('Values')

# Show plot
plt.show()
```

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create scatter plot
plt.scatter(x, y, color='blue')

# Annotate each point
for xi, yi in zip(x, y):
    plt.annotate(f'({xi}, {yi})', (xi, yi), textcoords="offset points", xytext=(5,5), ha='center')

# Set title and labels
plt.title('Scatter Plot with Annotations')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show plot
plt.show()
```

```python
# TASK: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.
import matplotlib.pyplot as plt

# Data
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

# Equal aspect ratio ensures the pie is drawn as a circle
plt.axis('equal')

# Set title
plt.title('Programming Language Usage')

# Show plot
plt.show()
```

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y_line = [1, 4, 9, 16, 25]
y_bar = [5, 7, 3, 8, 6]
y_scatter = [2, 4, 6, 8, 10]
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line chart
axs[0, 0].plot(x, y_line, marker='o')
axs[0, 0].set_title('Line Chart')

# Bar chart
axs[0, 1].bar(x, y_bar, color='skyblue')
axs[0, 1].set_title('Bar Chart')

# Scatter plot
axs[1, 0].scatter(x, y_scatter, color='green')
axs[1, 0].set_title('Scatter Plot')

# Pie chart
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
axs[1, 1].set_title('Pie Chart')
axs[1, 1].axis('equal')

# Layout adjustment
plt.tight_layout()
plt.show()
```

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
import matplotlib.pyplot as plt

# Data
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

# Create histogram
plt.hist(data, bins=5, color='purple', edgecolor='black')

# Add title and axis labels
plt.title('Histogram of Data Values')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show plot
plt.show()
```
